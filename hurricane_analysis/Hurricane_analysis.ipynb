```python
'''
This function takes in a list of damage amounts in dollars 
and converts M and B to millions and billions respectively.
It will leave entries where its 'Damages not recorded' as zero
'''

def cleanDamagesData(damages_list):
    print(damages_list)
    for index in range(len(damages_list)):
        value = damages_list[index]
        if(value == 'Damages not recorded'):
            print("skip")
        elif(value.find("M") > 0):
            value = (float(value.strip("M"))) * conversion["M"]
        else:
            value = (float(value.strip("B"))) * conversion["B"]
        damages_list[index] = value
    return damages_list
            
new_damages = cleanDamagesData(damages)
print(new_damages)
```

```python
'''
This function takes in lists of names, months, years, max_sustained_winds, areas_affected, damages(updated) and deaths
It then creates a dictionary for the items within these lists, they are in order so the first entry in names
corresponds with the first entry in months and so on
'''

def hurricaneDictionary(names, months, years, max_sustained_winds, areas_affected, new_damages, deaths):
    return {names[index]:{"Name":names[index],"Month": months[index],"Year": years[index],"Max Sustained Wind": max_sustained_winds[index],"Areas Affected": areas_affected[index],"Damage": new_damages[index],"Deaths": deaths[index]}  for index in range(len(names))}

hurricane_dict_byname = hurricaneDictionary(names, months, years, max_sustained_winds, areas_affected, new_damages, deaths)
print(hurricane_dict_byname)
```

```python
'''
This function takes a dictionary of hurricanes by name and outputs all hurricanes with the key now being 
the year they occured
'''

def hurricaneDictionaryByYear(hurricane_dict_byname):
    hurricanes_by_year = {}
    for key,value in hurricane_dict_byname.items():
        if(hurricanes_by_year.get(value['Year'])):
            hurricanes_by_year[value['Year']].append(value)
        else:
            hurricanes_by_year[value['Year']] = [value]
    return hurricanes_by_year

hurricaneDictionaryByYear(hurricane_dict_byname)
```

```python
'''
This function takes a dictionary of hurricanes by name and outputs the areas that get hit and how many times
'''
areas_affected_dict = {}
def hurricaneAreas(hurricane_dict_byname):
    areas_affected_dict = {}
    for hurricane in hurricane_dict_byname.values():
        for area in hurricane['Areas Affected']:
            if(areas_affected_dict.get(area)):
                areas_affected_dict[area]+=1
            else:
                areas_affected_dict[area]=1
    return areas_affected_dict
    
areas_affected_dict = hurricaneAreas(hurricane_dict_byname) 
areas_affected_dict
```

```python
'''
This function takes in the the hurricane dictionany with the names as keys and outputs the 
area that got hit the most by all hurricanes
'''
def areaMostHit(areas_affected_dict):
    return max(areas_affected_dict, key=areas_affected_dict.get)
mostHitKey = areaMostHit(areas_affected_dict)

print(mostHitKey + " is the most hit with " + str(areas_affected_dict[mostHitKey]) + " occurences")


```

```python
'''
This function takes in the the hurricane dictionany with the names as keys and outputs the 
hurricane by the most deaths. It outputs both the hurricane and the deaths that it caused. 
'''

def hurricanesMostDeaths(hurricane_dict_byname):
    deaths = 0
    hurricane = ""
    for key,value in hurricane_dict_byname.items():
        if value["Deaths"] > deaths:
            hurricane = key
            deaths = value["Deaths"]
    return hurricane,deaths

hurricane,deaths = hurricanesMostDeaths(hurricane_dict_byname)
print (hurricane + " had the most deaths with " + str(deaths))
```

```python
'''
This function takes in a number of deaths and outputs a mortality scale. This scale outputs a key based on the 
number of deaths as follows:

0: 0
1: 100
2: 500
3: 1000
4: 10000

'''

def getMortalityRating(deaths):
    selected_key = 0
    for key,death_count in mortality_scale.items():
        if(deaths >= death_count):
            selected_key = key
            continue
        else:
            return selected_key
'''
This function takes in the the hurricane dictionany with the names as keys and outputs all the hurricane's based
on their mortality rate
'''

def hurricanesByMortality(hurricane_dict_byname):
    hurricane_mortality_dict = {}
    for name,value in hurricane_dict_byname.items():
        if(hurricane_mortality_dict.get(getMortalityRating(value["Deaths"]))): 
            hurricane_mortality_dict[getMortalityRating(value["Deaths"])].append(name)
        else:
            hurricane_mortality_dict[getMortalityRating(value["Deaths"])] = [name]
    return hurricane_mortality_dict
        
print(hurricanesByMortality(hurricane_dict_byname))


```

```python
'''
This function takes in the the hurricanes plus the damages list and combines them together
'''
def createDictHurricanesByDamage(names,new_damages):
    return dict(zip(names,damages))

'''
This function takes in the the hurricanes combined with the damage the cost in dollars
It removes not float/int values,sorts the dictionary with largest first and 
returns the sorted dictionary plus the hurricane with the largest damage and the damage it causes
'''

def getHurricaneLargestDamage(damages_by_hurricane_dict):
    
    damages_by_hurricane_dict = {key:val for key, val in damages_by_hurricane_dict.items() if val != 'Damages not recorded'}
    damages_by_hurricane_dict_sorted = dict(sorted(damages_by_hurricane_dict.items(), key=lambda item: item[1],reverse=True))
    return damages_by_hurricane_dict_sorted, list(damages_by_hurricane_dict_sorted.keys())[0], list(damages_by_hurricane_dict_sorted.values())[0]

damages_by_hurricane_dict = createDictHurricanesByDamage(names,new_damages)
hurricane_dict_bydamages_sorted, hurricane_most_damages, damage_cost = getHurricaneLargestDamage(damages_by_hurricane_dict)
print("Hurricane with the most damage is " + hurricane_most_damages + " with the following damage US$" + str(damage_cost))
```

```python
# 9
# Rating Hurricanes by Damage
damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}


'''
This function takes in a damage cost and returns the appropriate scale based on
0: US$0
1: US$100000000
2: US$1000000000
3: US$10000000000
4: US$50000000000
'''
def getDamageScaleByDamageCost(damage):
    selected_key = 0
    for key,damage_cost in damage_scale.items():
        #print("Checking if the following damage " + str(damage) + " is in this range " + str(damage_cost) + " with key " + str(key))
        if(damage_cost <= damage):
            selected_key = key        
    return selected_key 
        
'''
This function takes in the the hurricanes sorted by damage and outputs these hurricanes on a damage scale
'''

def hurricanesByDamageScale(hurricane_dict_bydamages_sorted):
    hurricane_damage_dict = {0:[],1:[], 2:[],3:[],4:[]}
    for name,damage in hurricane_dict_bydamages_sorted.items():
        damage_scale = getDamageScaleByDamageCost(damage)
        #print(str(damage) + " for Damage scale = " + str(damage_scale))
        hurricane_damage_dict[damage_scale].append(name)
    return hurricane_damage_dict

print(hurricanesByDamageScale(hurricane_dict_bydamages_sorted))
```
